```python
git status     #This command is used to display the state of the working directory and the staging area. It allows you to see which changes have been staged, which haven't, and which files aren?t being tracked by Git. It also lists the files that you've changed and those you still need to add or commit.
```

```python
git add .    #This command is used to add all the files present in our repository and save changes to the staging area.
```

```python
git commit -m "This is my first commit"    #The commit command performs a commit, and the -m "message" adds a message.
```
